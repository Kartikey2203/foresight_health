```python
product_revenue = (
    df.groupby("Product Name")["Sales"]
      .sum()
      .reset_index()
      .sort_values("Sales", ascending=False)
)

product_revenue["Revenue_%"] = (
    product_revenue["Sales"]
    / product_revenue["Sales"].sum()
    * 100
)

product_revenue["Cum_%"] = (
    product_revenue["Revenue_%"]
    .cumsum()
)

def abc_segment(x):
    if x <= 80:
        return "A"
    elif x <= 95:
        return "B"
    else:
        return "C"

product_revenue["Segment"] = (
    product_revenue["Cum_%"]
    .apply(abc_segment)
)

product_revenue.head()
```

```python
customer_sales = (
    df.groupby("Customer Name")["Sales"]
      .sum()
      .reset_index()
      .sort_values("Sales", ascending=False)
)

customer_sales["Revenue_%"] = (
    customer_sales["Sales"]
    / customer_sales["Sales"].sum()
    * 100
)

customer_sales["Cum_%"] = (
    customer_sales["Revenue_%"]
    .cumsum()
)

def abc_segment(x):
    if x <= 80:
        return "Platinum"
    elif x <= 95:
        return "Gold"
    else:
        return "Regular"


customer_sales["Segment"] = (
    customer_sales["Cum_%"]
    .apply(abc_segment)
)

customer_sales.to_csv("customer_sales.csv")
```

```python
rep_metrics = (
    df.groupby("Sales Rep")
      .agg({
          "Sales":"sum",
          "Quantity":"sum",
          "Customer Name":"nunique"
      })
      .reset_index()
)

rep_metrics.columns = [
    "Sales Rep",
    "Revenue",
    "Quantity",
    "Customers"
]
rep_metrics["Revenue_Score"] = (
    rep_metrics["Revenue"]
    / rep_metrics["Revenue"].max()
)

rep_metrics["Quantity_Score"] = (
    rep_metrics["Quantity"]
    / rep_metrics["Quantity"].max()
)

rep_metrics["Customer_Score"] = (
    rep_metrics["Customers"]
    / rep_metrics["Customers"].max()
)
```

```python
rep_metrics["Performance_Score"] = (
      0.5 * rep_metrics["Revenue_Score"]
    + 0.3 * rep_metrics["Quantity_Score"]
    + 0.2 * rep_metrics["Customer_Score"]
)

rep_metrics.sort_values(
    "Performance_Score",
    ascending=False
)
```

```python
month_order = {
    "January":1,
    "February":2,
    "March":3,
    "April":4,
    "May":5,
    "June":6,
    "July":7,
    "August":8,
    "September":9,
    "October":10,
    "November":11,
    "December":12
}

monthly_sales = (
    df.groupby(["Year","Month"])["Sales"]
      .sum()
      .reset_index()
)

monthly_sales["Month_Num"] = monthly_sales["Month"].map(month_order)

monthly_sales = monthly_sales.sort_values(
    ["Year","Month_Num"]
)

monthly_sales["MoM_Growth_%"] = (
    monthly_sales["Sales"]
    .pct_change()
    * 100
)

# monthly_sales
```

```python
forecast_df = (
    df.groupby(["Year","Month"])["Sales"]
      .sum()
      .reset_index()
)

forecast_df["ds"] = pd.to_datetime(
    forecast_df["Year"].astype(str)
    + "-"
    + forecast_df["Month"].astype(str)
    + "-01"
)

forecast_df.rename(
    columns={"Sales":"y"},
    inplace=True
)

forecast_df = forecast_df[["ds","y"]]
```

```python
from prophet import Prophet

model = Prophet()

model.fit(forecast_df)

future = model.make_future_dataframe(
    periods=12,
    freq="M"
)

forecast = model.predict(future)

forecast.tail()
```
